```python
df_years = World_population[["Continent", "1970 Population", "1980 Population", "1990 Population", 
              "2000 Population", "2010 Population", "2015 Population", 
               "2020 Population", "2022 Population"]]


df_long = df_years.melt(id_vars=["Continent"], 
                        var_name="Year", 
                        value_name="Population")


df_long["Year"] = df_long["Year"].str.extract(r'(\d{4})').astype(int)


df_continent = df_long.groupby(["Continent", "Year"], as_index=False)["Population"].sum()


fig = px.line(
    df_continent,
    x="Year", 
    y="Population", 
    color="Continent",  
    markers=True,
    title="Évolution de la population par continent (1970 - 2022)",
    labels={"Population": "Population Totale", "Year": "Année"}
)

fig.show()
```

###  Conclusions

- **L’Asie** reste le continent le plus peuplé.
- **L’Afrique** connaît la croissance démographique la plus rapide.
- **L’Europe** montre une tendance de stabilisation ou de légère baisse.
- La répartition de la population est **inégale** selon les régions, influencée par l’économie, la géographie et l’histoire.

##  Conclusion

L'analyse met en évidence des dynamiques démographiques contrastées selon les régions du monde :

- **L’Asie** demeure le continent le plus peuplé, soutenu par la présence de pays à très forte densité comme la Chine, l’Inde ou l’Indonésie.
- **L’Afrique** présente **la croissance démographique la plus rapide**, portée par un taux de natalité élevé et une population majoritairement jeune. Elle jouera un rôle démographique central dans les prochaines décennies.
- **L’Europe** est caractérisée par une **stagnation voire un déclin de la population**, lié à une natalité faible et à un vieillissement généralisé.
- **L’Amérique** affiche des dynamiques variées : croissance modérée en Amérique du Nord et évolution plus contrastée en Amérique latine.
- La répartition mondiale de la population est fortement influencée par :
  - les **conditions économiques** (niveau de développement, emploi),
  - les **facteurs historiques** (colonisation, migrations),
  - les **facteurs géographiques** (climat, ressources, urbanisation).

###  Insight clé
La population mondiale continue de croître, mais **de manière inégale** : tandis que certains continents voient leur population augmenter rapidement, d’autres entrent dans une phase de vieillissement et de décroissance. Ces évolutions auront des **impacts majeurs** sur les politiques économiques, environnementales, éducatives et migratoires des prochaines années.
